## 1. Handling missing variables

```python
missing = df.isnull().sum()
missing = missing[missing>0]
missing
```

Two columns have missing values, Loan End Date and Hits

### 1.1 Handling Loan Date End column for missing values

```python
# This shows the rows where 'Loan Date End' have values, so we can see what exactly these values are.

df.loc[df['Loan Date End'].notnull()].head()
```

From the result above, we notice that Loan date end is similar to the contract column. So we will get the year from contract columns and use them in the Loan Date End column. 
However I will be creating a new column all togeter called Contract expiry year and will have the latter year from contract column.
Then I will drop both contract and Laon Date End column


```python
# Create a new column which contains the split values and remove white space in Loan date end,
# then drop the month column

df[['month','Contract Expiry Year']] = df['Loan Date End'].str.split(',', expand = True).apply(lambda x: x.str.strip())
df.drop(['month','Loan Date End'] , axis =1, inplace = True)
df.head()
```

```python
# Get the last 4 characters of each value in the contract column and fill the null values of Contract Expiry Year 
# with this result

lastFourchars = df['Contract'].apply(lambda x : x[-4:])
df['Contract Expiry Year'] = df['Contract Expiry Year'].fillna(lastFourchars)
df.head()
```

```python
# Drop contract column

df.drop('Contract', axis = 1, inplace = True)
```

```python
df.head()
```

```python
df['Contract Expiry Year'].isnull().sum()
```

NOTE: We have handled all the null values in the former Loan Date End column, now known as Contract Expiry Year 

### 1.2 Handling Hit column for missing values

Hits(means the number of times a player profile has been viewed)- we can set this to 0 meaning the player has no view.

```python
df['Hits'] = df['Hits'].fillna(0)
df['Hits'].isnull().sum()
```

NOTE: We have handled all the null values for Hits column
